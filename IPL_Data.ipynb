```python
### Load required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the IPL dataset
data = pd.read_csv('E:\SYMBIOSIS MSC DSSA\ipl_dataset.csv')
ipl= pd.DataFrame(data)
ipl
```

```python
# Have a glance at first 10 rows of the dataset

ipl.head(10)
```

```python
# Find the number of rows and columns

print('The number of rows are ',ipl.shape[0])
print('The number of columns are ',ipl.shape[1])
```

```python
# Get the frequency of most 'Man of the Match' awards

count=ipl['player_of_match'].value_counts()
count
```

```python
# Get Top 10 players with most man of the matches

count[:10]   # or you can also write--  ipl['player_of_match'].value_counts()[:10]
```

```python
# Get Top 5 players with most man of the matches

ipl['player_of_match'].value_counts()[:5]
```

```python
# bar plot on the top 5 man of the matches and how many won

players = list(ipl['player_of_match'].value_counts()[:5].keys())
winnings = list(ipl['player_of_match'].value_counts()[:5])
plt.bar(players,winnings,color='orange',edgecolor='k')
plt.xlabel('Player Names')
plt.ylabel('Number of Man of the Match Titles won')
plt.title('MAN OF THE MATCHES')
plt.show()
```

```python
# count the result column

ipl['result'].value_counts()
```

```python
# find the number of toss wins with respect to each team

ipl['toss_winner'].value_counts()
```

```python
# extract records where a team won by batting first

winner_batting_first = ipl[ipl['win_by_runs'] != 0]
winner_batting_first
```

```python
# to determine margin of runs while winning of the teams

plt.hist(winner_batting_first['win_by_runs'],color='magenta',edgecolor='k')
plt.xlabel('Runs')
plt.title('Distribution of Runs')
plt.grid(True)
plt.show()
```

```python
# finding out the number of wins w.r.t each team after batting first

first_batting = ipl[ipl['win_by_runs'] != 0]
first_batting['winner'].value_counts()
```

```python
# plotting a bar plot for top 3 teams with most wins after batting first

top_3 = first_batting['winner'].value_counts().head(3)
plt.bar(top_3.keys(), top_3, color = ['r','b','g'], edgecolor = 'k')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Top 3 Teams with Most Wins')
plt.show()
```

```python
# making a pie_chart

plt.figure(figsize = (7,7))
labels = list(first_batting['winner'].value_counts().keys())
plt.pie(list(first_batting['winner'].value_counts()), labels = labels , autopct = '%0.2f%%')
plt.show()
```

```python
# looking at the number of matches played each season

ipl['season'].value_counts()
```

```python
# looking at the number of matches played in each city

ipl['city'].value_counts()
```

```python
# finding out records where a team won the match after batting second

batting_second = ipl[ipl['win_by_wickets'] != 0]
batting_second
```

```python
# extract records where team won both the toss and the match

first_batting_won = ipl[ipl['winner'] == ipl['toss_winner']]
first_batting_won
```

```python
# plot a histogram to show number of wins w.r.t number of wickets

plt.hist(batting_second['win_by_wickets'],edgecolor = 'k',color='g')
plt.xlabel('Number of Wickets')
plt.ylabel('Number of Wins')
plt.title('Number of wins w.r.t number of wickets')
plt.show()
```

```python
# frequency of number of wins after batting second

second_batting = ipl[ipl['win_by_wickets'] != 0]
second_batting_wins = second_batting['winner']
second_batting_wins.value_counts()
```

```python
# finding out how many times a team won match after winning toss

toss_and_match_win = ipl[ipl['toss_winner'] == ipl['winner']]
toss_and_match_win['winner'].value_counts()
```
